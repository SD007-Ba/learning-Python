# **Coclusion**
 This analysis of the Titanic dataset provides several key insights into the survival patterns of passengers:

	-	Gender and Survival: Female passengers had a significantly higher survival rate compared to male passengers, suggesting that gender played a major role in survival.
	-	Passenger Class: First-class passengers had better chances of survival than those in second and third class. This indicates a strong link between socioeconomic status and survival probability.
	-	Age Factor: Younger passengers, especially children, had a higher likelihood of survival compared to older adults. This aligns with the “women and children first” evacuation approach.
	-	Family Influence: Passengers traveling with family (i.e., with siblings/spouses or parents/children) had a slightly better survival rate than those traveling alone.
	- Fare Impact: Passengers who paid higher fares (often first-class) had a better survival rate, reinforcing the influence of class.
	

Overall, the data suggests that survival was influenced by a combination of gender, age, class, and family connections. These findings reflect the social dynamics and evacuation priorities of the time.
